## 1.等值线图
### 具体解释看书本+自行上网搜集

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dx = 0.01;dy = 0.01
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = np.meshgrid(x,y)
```

<https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html#numpy.meshgrid>

```python
def f(x,y):
    return (1-y**5+x**5)*np.exp(-x**2-y**2)
```

```python
X
```

```python
z = plt.contour(X,Y,f(X,Y),8,colors='black')
```

> cpuntour<https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.contour.html>

```python

plt.contourf(X,Y,f(X,Y),8)
plt.clabel(z,inline=1,fontsize=10)
```

# 等势线添加颜色组合：contourf()设置-cmap参数

```python
#使用之前的数据
plt.contourf(X,Y,f(X,Y),8,
            cmap=plt.cm.hot)
plt.colorbar()
```

# 极区图
## bar()函数绘制，input：角度 $\theta$、半径

```python
N = 8
theta = np.arange(0.,2*np.pi,2*np.pi/N)
radii = np.array([4,7,5,3,1,5,6,7])
```

```python
plt.axes([0.025,0.025,0.95,0.95],polar=True)
colors = np.array(['#4bb2c5','#c5b47f','#eaa228','#579575','#839577','#958c12','#953579','#4b5de4',])
```

# 自己傻乎乎的输入半天的自定义颜色···

>axes <https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.axes.html>

```python
bars = plt.bar(theta,radii,width=(2*np.pi/N),bottom=0.0,color=colors)
```

```python
plt.axes([0.025,0.025,0.95,0.95],polar=True)
bars = plt.bar(theta,radii,width=(2*np.pi/N),bottom=0.0,color=colors)
```

# 极区图颜色：#rrggbb——>名称

```python
color = np.array(['lightgreen','darkred','navy','brown','violet','plum','yellow','darkgreen'])
```

```python
plt.axes([0.025,0.025,0.95,0.95],polar=True)
bars = plt.bar(theta,radii,width=(2*np.pi/N),bottom=0.0,color=color)
plt.savefig('pie.EPS')
```
