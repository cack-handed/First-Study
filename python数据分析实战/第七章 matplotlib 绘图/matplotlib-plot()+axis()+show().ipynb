# matplitlib绘图优点
* 使用极其简单
* 渐进、交换式方式实现数据可视化
* 表达式和文本使用LaTex排版
* 对图像元素控制力更强
* 可输出PNG、PDF、SVG和EPS等多种格式
**常见的矢量图包括：EPS，EMF和SVG。SVG适合于浏览器，EPS适合于LaTeX，EMF适合于Word**

   # matplotlib 架构
   逻辑上来讲，matplotlib整体分为三层：   
   ## scripting(脚本层)-->artist(表现层)-->backend(后端层)
   **层与层之间只能向下通信，而不能向上通信**

 ***
## backend层
* FigureCanvas对象实现了绘图区域的概念
* Renderer对象在FigureCanvas上绘图
* Event对象处理用户输入（键盘鼠标事件）
***
## Artist层
*图形中所有能看到的元素都属于Artist对象
* 标题：title
* 轴标签：x-label
* 刻度：ticks
* 单条轴：axis 负责展示在axes对象上的数值，**定义数值范围，管理刻度（轴上的标记）和刻度值标签（代表每个刻度大小的文本标签）**。刻度的位置用**Locator**对象调整，刻度标签的格式用**formatter**对象调整
* axes ：通常表示图形或图表式对什么内容进行作图的。每个axes只属于一个figure。有两个（三维就有两个）artist axis组成。**标题、x标签、y标签**等都属于axes这个composite artist类型的对象
* 图形：figure **在artist层的最上面，对应整个图形表示，通常可包含多条轴（axes）**
#### Artist类分为两种：原始（primitive）复合（composite）
***

```python
import math
import numpy as np
t = np.arange(0,2.5,0.1) #t--x轴范围0，2.5，步长0.1
y1 = np.sin(math.pi*t)  
y2 = np.sin(math.pi*t+math.pi/2)
y3 = np.sin(math.pi*t-math.pi/2) 
```

```python
plt.plot(t,y1,'b*')
```

```python
plt.plot(t,y2,'g^')
```

```python
plt.plot(t,y3,'ys')
```

```python
plt.plot(t,y1,'b--',t,y2,'g',t,y3,'r--')
```
