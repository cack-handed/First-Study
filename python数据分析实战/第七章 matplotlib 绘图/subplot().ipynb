# kwargs
组成图表的各个对象有很多用以描述他们特点的属性，这些属性均有默认值，但是可以使用**关键字参数（keyword args，简介kwargs）** 设置
***
matplotlib.pyplot.plot(\*args,\**kwargs)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,4,2,1,0,1,2,1,4],linewidth=2.0) #linewidth 就是关键字参数
```

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)
#y1
```

```python
plt.subplot(2,1,1)
plt.plot(t,y1,'b-.')
plt.subplot(2,1,2)
plt.plot(t,y2,'r--')
```

```python
t = np.arange(0.,1.,0.05)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'b-.')
plt.subplot(1,2,2)
plt.plot(t,y2,'r--')
```
