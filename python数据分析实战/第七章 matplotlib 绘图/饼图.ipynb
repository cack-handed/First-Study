# plt.pip函数
***
## 基础图
### pie参数中起始角度设置-startangel

```python
import matplotlib.pyplot as plt
```

```python
labels = ['Nokia','Samsung','Apple','Lumia']
values = [10,30,45,15]
colors = ['yellow','green','red','blue']
```

```python
plt.pie(values,labels=labels,colors=colors)
plt.axis('equal')
```

# 强调—explode

```python
explode=[0.3,0,0,0]
plt.title=('A pie Chart')
plt.pie(values,labels=labels,colors=colors,explode=explode,startangle=180)
plt.axis=('equal')
```

# pie参数中加入百分比-autopct
# pie参数中加入阴影-shadow

```python
plt.pie(values,labels=labels,colors=colors,explode=explode,
        shadow=True,
       autopct='%1.1f%%',startangle=180)
plt.axis=('equal')
```

# DataFrame数据绘图
## 每幅图只能表示一个序列
## DataFrame- plot()绘图种类参数设置-kind

```python
import pandas as pd
```

```python
data = {'series1':[1,3,4,3,5],
       'series2':[2,4,5,2,4],
       'series3':[3,2,3,1,3]}
df = pd.DataFrame(data)
df['series1'].plot(kind='pie',figsize=(10,5),explode=[0.3,0,0,0,0],autopct='%1.1f%%')
```
