# pyplot用于绘制直方图的函数为hist()
该函数具有一个其他绘图函数没有的功能。   
**除了绘制直方图之外，还可以以元祖形式返回直方图的计算结果(n,bins,patches)**

```python
import matplotlib.pyplot as plt
import numpy as np
pop = np.random.randint(0,100,100)
pop
```

```python
n,bins,patches = plt.hist(pop,bins=20)
```

# 条状图

```python
import matplotlib.pyplot as plt
index = [0,1,2,3,4]
values = [5,7,3,4,6]
plt.bar(index,values)
```

## x轴显示类别的简单条状图

```python
plt.bar(index,values)
plt.xticks(index,['A','B','C','D','E'])
```

```python
std1 = [0.8,1,0.4,0.9,1.3]
plt.title('A Bar Chart')
# error_kw属性是误差
plt.bar(index,values,yerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='first')
plt.xticks(index,['A','B','C','D','E'])
plt.legend(loc=2)
```

# 水平条状图


```python
##### 使用之前的数据
plt.title('A Horizontal Bar Chart')
plt.barh(index,values,xerr=std1,error_kw={'ecolor':'0.1','capsize':6},
         alpha=0.3,label='first')
plt.yticks(index,['A','B','C','D','E'])
plt.legend(loc=5)
```

# 多序列条状图

```python
values1=[5,7,3,4,6]
values2=[6,6,4,5,7]
values3=[5,6,5,4,6]
bw = 0.3
plt.axis([0,5,0,8])
plt.title('A Multiseries Bar Chart',fontsize=20)
plt.bar(index,values1,bw,color='b')
plt.bar(index,values2,bw,color='g')
plt.bar(index,values3,bw,color='r')
plt.xticks(index,['A','B','C','D','E'])
```

```python
import pandas as pd
data = {'series1':[1,3,4,3,5],
       'series2':[2,4,5,2,4],
       'series3':[3,2,3,1,4]}
df = pd.DataFrame(data)
df.plot(kind='bar')
```

# 多序列堆积条状图

```python
# use before data
import numpy as np
plt.axis([0,15,-0.5,3.5])
plt.title('A Multiseries Horizontal Stacked Bar Chart')
plt.barh(index,data['series1'],color='w',hatch='xx')
plt.barh(index,data['series2'],color='w',hatch='///',left=data['series1'])
plt.barh(index,data['series3'],color='w',hatch='\\\\\\',left=np.array(data['series1'])+np.array(data['series2']))
plt.yticks(index,['Jan18','Feb18','Mar18','Apr18'])
```


# pandas DataFrame绘制堆积条状图

```python
df = pd.DataFrame(data)
df.plot(kind='bar',stacked=True)
```

# 其他条状图

```python
import matplotlib.pyplot as plt
import numpy as np
x0 = np.arange(8)
y1 = np.array([1,3,4,6,5,3,2,1])
y2 = np.array([1,2,5,4,3,3,2,1])
plt.ylim(-7,7)
plt.bar(x0,y1,0.9,facecolor='r')
plt.bar(x0,-y2,0.3,facecolor='b')
plt.xticks(())
plt.grid(True)
for x,y in zip(x0,y1):
    plt.text(x+0.4,y+0.05,'%d' %y,ha='right',va='bottom')
for x,y in zip(x0,y2):
    plt.text(x+0.4,-y-0.05,'%d' %y,ha='center',va='top')
```
