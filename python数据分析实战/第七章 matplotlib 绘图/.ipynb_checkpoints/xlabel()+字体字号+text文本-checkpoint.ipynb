```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.axis([0,5,0,20])
plt.title('My first plot')
plt.xlabel('Counting')
plt.ylabel('Square values')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

# xlabel

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting')
plt.ylabel('Square values')

plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

# 字体字号+文本

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting')
plt.ylabel('Square values')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
```

# latex公式，字号，背景，透明度

```python
plt.axis([0,5,0,20])
plt.title('My first plot',fontsize=20,fontname='Times New Roman')
plt.xlabel('Counting')
plt.ylabel('Square values')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
# facecolor 背景颜色，alpha透明度
plt.text(1.1,12,r'$y = x^2$',fontsize=20,bbox={'facecolor':'yellow','alpha':0.3}) 
```
