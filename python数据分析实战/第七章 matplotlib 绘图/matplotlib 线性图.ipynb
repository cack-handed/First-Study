# 如何绘制线性图
* 确定两条轴的刻度范围
* 设定两条轴的名称
* 传入x，y进行描点   
 ***根据之前所讲的自主选择text，label，legend，save等操作***

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.05)
y = np.sin(3*x)/x
plt.plot(x,y)
```

```python
y2 = np.sin(2*x)/x
y3 = np.sin(3*x)/x/2
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.legend(['y1','y2','y3'])
```

```python
# 使用之前的x，y值
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y2,'m-.')
plt.plot(x,y3,color='#87a3cc',linestyle='--')
plt.legend(['y1','y2','y3'])
```

# 调整轴的刻度

```python
# 使用之前的x，y值
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y2,'m-.')
plt.plot(x,y3,color='#87a3cc',linestyle='--')
plt.legend(['y1','y2','y3'])
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],
          [r'$-2\pi$','$-\pi$','$0$','$\pi$','$2\pi$'])
plt.yticks([-1,0,1,2,3],
          ['$-1$','$0$','$+1$','$+2$','$+3$'])
```

# 笛卡尔坐标系

```python
# 使用之前的x，y值
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y2,'m-.')
plt.plot(x,y3,color='#87a3cc',linestyle='--')
plt.legend(['y1','y2','y3'])
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],
          [r'$-2\pi$','$-\pi$','$0$','$\pi$','$2\pi$'])
plt.yticks([-1,0,1,2,3],
          ['$-1$','$0$','$+1$','$+2$','$+3$'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

# 添加注释和箭头

```python
# 使用之前的x，y值
plt.plot(x,y,'k--',linewidth=3)
plt.plot(x,y2,'m-.')
plt.plot(x,y3,color='#87a3cc',linestyle='--')
plt.legend(['y1','y2','y3'])
plt.xticks([-2*np.pi,-np.pi,0,np.pi,2*np.pi],
          [r'$-2\pi$','$-\pi$','$0$','$\pi$','$2\pi$'])
plt.yticks([-1,0,1,2,3],
          ['$-1$','$0$','$+1$','$+2$','$+3$'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x}=1$',xy=[0,1],xycoords='data',
            xytext=[60,60],fontsize=16,textcoords='offset points',
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.5'))
```
